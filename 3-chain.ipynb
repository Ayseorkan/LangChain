```python
from langchain_core.prompts import PromptTemplate
template = "Tell me a {adjective} joke about {content}."
prompt = PromptTemplate(
    input_variables=["adjective", "content"],  
    template=template,
)
formatted_prompt = prompt.format(adjective="funny", content="chickens")
print(formatted_prompt)
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(temperature=0)
response = llm.invoke("Give me a suggestion for today's lunch.")
print(response.content)
```

```python
from langchain_core.output_parsers import StrOutputParser
outout_parser = StrOutputParser()
```

```python
chain = prompt | llm | outout_parser
result = chain.invoke({"adjective": "hilarious", "content": "computers"})
print(result)
```
